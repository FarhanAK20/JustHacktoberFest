# Matplotlib

## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  
```

%matplotlib inline is used to avoid pop-ups & show graphs inline with the code

```python
x=[i for i in range(0,5)]
x
```

```python
y=[i**3 for i in range(0,5)]
y[0]=0
y
```

```python
plt.plot(x,y)
plt.xlabel("This is x-label")
plt.ylabel("This is y-label")
plt.title("Title")
plt.show()
```

```python
plt.plot(x,y)
plt.xlim([0,2])
plt.ylim([0,10])
plt.show()
```

You'll observe that the cell will still give an output but with an "out cell". This line is necesarry if you're using anything other than Jupyter notebook.

```python
plt.plot(x, y, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)
# plt.plot(x, y, 'go--', linewidth=2, markersize=12)
plt.show()
```

Checkout the documentation for matplotlib.pyplot.plot (plt.plot here) here,<br>
https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html

```python
x=[i for i in range(101)]
y1=[i**2 for i in x]
y2=[i**4 for i in x]
```

```python
fig=plt.figure(figsize=(12,5))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

ax1.plot(x,y1)
ax1.set_title("Title for 1")
ax1.set_xlabel("X LABEL")
ax1.set_ylabel("Y LABEL")

ax2.plot(x,y2)
ax2.set_title("Title for 2")
# plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.plot(x,y1,'r')
plt.title("x*2")

plt.subplot(2,2,2)
plt.plot(x,y2,'b')
plt.title("x*4")

plt.subplot(2,2,3)
plt.plot(x,y2,'b')
plt.title("2nd x*4")

plt.subplot(2,2,4)
plt.plot(x,y2,'r')
plt.title("x*4")

plt.show()
```

```python
y1=[i//2 for i in y1]
```

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,1,1)

ax1.plot(x,y1,label="X**4")
ax1.plot(x,y2,label="X**2")
ax1.legend(loc=0)
plt.show()
# ax1.legend(loc=(0.5,0.7))
```

Documentation for plt.legend() <br>
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,1,1)

ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.legend(["X**4","X**2"],loc=0)
plt.show()
```

# Seaborn

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue="sex",palette='BuGn_r')
```

```python
sns.distplot(tips['total_bill'],bins=10)
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

Check documentation here.<br>
https://seaborn.pydata.org/generated/seaborn.scatterplot.htm

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind='hex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

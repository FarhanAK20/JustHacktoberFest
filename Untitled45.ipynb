```python
def calculate_bValue(data,amount_column_name):
    """
        :param data: data frame of data for which feature is to be calculated with a and delay_avg present.
        :param amount_column_name:  name of amount column.

        """
    data['b'] = np.where((data['a'] > 1) & (data['delay_avg'] != 0),data[amount_column_name] / data['delay_avg'],data['delay_avg'] / data[amount_column_name])
    return (data['b'].tolist())
```
